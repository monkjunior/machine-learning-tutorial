```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_square.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
```

```python
x = np.hstack((np.ones((N, 1)), x, x*x))
w = np.array([2000., -100., 1.]).reshape(-1,1)
numOfIteration = 300
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000000002

print(w[0][0])
print(w[1][0])
print(w[2][0])
print(x[0:5, :])
print(y[0:5, :])
```

```python
for i in range(1, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
    w[2] -= learning_rate*np.sum(np.multiply(r, x[:,2].reshape(-1,1)))
    print(cost[i])
```

```python
w_2 = w[2][0]
w_1 = w[1][0]
w_0 = w[0][0]
xs = np.linspace(-25, 100, 1000)
ys = w_2*xs*xs + w_1*xs + w_0
plt.scatter(data[:, 0].reshape(-1, 1), data[:, 1].reshape(-1, 1))
plt.xlabel('mét vuông')
plt.ylabel('giá')
plt.plot(xs, ys, '-r', label=f'y = {round(w_2, 2)}x**2 + {round(w_1, 2)}x + {round(w_0, 2)}')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
